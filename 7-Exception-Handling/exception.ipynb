```python
## exception try,except,else,finally
try:
    a=b;
except NameError as ex:
    print(ex)
else:
    print("no error")
finally:
    print("finally")
```

```python
try:
    result=10/0
except ZeroDivisionError as ex:
    print(ex)
    print("please enter valid value")
except Exception as ex1:
    print(ex1)
else:
    print("no error")
finally:
    print("finally")
```
