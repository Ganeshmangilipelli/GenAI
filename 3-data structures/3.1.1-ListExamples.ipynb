```python
## example of organizing the student grades in a list

grades = [90, 80, 70, 60]
total = 0
for grade in grades:
    total += grade
average = total / len(grades)
print("Average grade:", average)
```

```python
## managing the inventory using the list real life example

inventory = ["shoes", "socks", "pants", "shirts"]
for item in inventory:
    print("item:", item)

if "shoes" in inventory:
    print("shoes are in the inventory")
else:
    print("shoes are not in the inventory")

if "shirts" not in inventory:
    print("shirts are not in the inventory")
else:
    print("shirts are in the inventory")
```

```python
##example of collecting user feedback using a list real life example

feedback=["great service","poor service","average service","good service","excellent service"]
feedback.append("poor service")
positive=sum(1 for f in feedback if f.startswith("great") or f.startswith("good") or f.startswith("excellent"))
negative=sum(1 for f in feedback if f.startswith("poor") or f.startswith("average"))
print("positive feedback:",positive)
print("negative feedback:",negative)

```
