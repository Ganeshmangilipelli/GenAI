```python
## accessing elements in a tuple
numbers = (1, 2, 3, 4, 5)
print(numbers[0])  # Accessing the first element
print(numbers[-1])  # Accessing the last element
print(numbers[0:4])  # Accessing a range of elements
print(numbers[::])  # Accessing all elements
```

```python
## immutable nature of tuples
## tuples are immutable, meaning their elements cannot be changed after creation.
list1 = [1, 2, 3]
print("Before modification:", list1)
list1[0] = 10  # Modifying the first element
print("After modification:", list1)

tuple1 = (1, 2, 3)
print("Before modification:", tuple1)
# tuple1[0] = 10  # This will raise a TypeError
```

```python
## tuple methods
print(numbers.count(2))  # Count occurrences of 2
print(numbers.index(3))  # Index of the first occurrence of 3
print(len(numbers))  # Length of the tuple
print(min(numbers))  # Minimum value in the tuple
print(max(numbers))  # Maximum value in the tuple
```

```python
## packing and unpacking tuples
# Packing a tuple
packed_tuple = 1, 2, 3, "Hello"
print("Packed Tuple:", packed_tuple)

# Unpacking a tuple
a, b, c, d = packed_tuple
print("Unpacked Tuple:", a)
print("Unpacked Tuple:", b)
print("Unpacked Tuple:", c)
print("Unpacked Tuple:", d)
```

```python
## unpacking with asterisk
numbers = (1, 2, 3, 4, 5)
a, *b, c = numbers
print(a)  # 1
print(b)  # [2, 3, 4]
print(c)  # 5
```

```python
nested_tuple = (1, 2, (3, 4), (5, 6))
print(nested_tuple[2])  # Accessing the third element, which is a tuple
print(nested_tuple[2][1])  # Accessing the second element of the third tuple
```

```python
## iterating through a nested tuple
nested_tuple = ((1, 2), (3, 4), (5, 6))
for element in nested_tuple:
    for item in element:
        print(item,end=" ")
    print()
    
```
