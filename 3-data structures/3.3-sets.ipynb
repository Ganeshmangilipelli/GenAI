```python
## sets in python

## sets are collection of unique elements
## sets are unordered
## sets are mutable
## sets are indexed
## sets are iterable

## sets are used to store unique elements
## sets are used to remove duplicates
```

```python
##creating a set
set1 = {1, 2, 3, 4, 5}
print(set1)
print(type(set1))

##creating a set from a list
list1 = [1, 2, 3, 4, 5]
set2 = set(list1)
print(set2)

##creating a set from a tuple
tuple1 = (1, 2, 3, 4, 5)
set3 = set(tuple1)
print(set3)

##creating a set from a string
string1 = "Hello"
set4 = set(string1)
print(set4)
```

```python
## set operations
setA = {1, 2, 3, 4, 5}
setB = {4, 5, 6, 7, 8}
print(setA.union(setB))
print(setA.intersection(setB))
print(setA.difference(setB))
print(setB.difference(setA))
print(setA.symmetric_difference(setB))
print(setA.issubset(setB))
print(setA.issuperset(setB))
```

```python
## set methods
setA = {1, 2, 3, 4, 5}
setA.add(6)
print(setA)
setA.remove(3)
print(setA)
setA.discard(4)
print(setA)
setA.pop()
print(setA)
setA.clear()
print(setA)
```
