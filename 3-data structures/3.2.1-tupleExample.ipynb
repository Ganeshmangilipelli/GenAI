```python
#real life example of tuples

##example of collecting user feedback using a tuple real life example

feedback=("great service","poor service","average service","good service","excellent service")
positive=sum(1 for f in feedback if f.startswith("great") or f.startswith("good") or f.startswith("excellent"))
negative=sum(1 for f in feedback if f.startswith("poor") or f.startswith("average"))
print("positive feedback:",positive)
print("negative feedback:",negative)

```
