```python
##arithmetic operations
a=10
b=10
add_result=a+b
sub_result=a-b
mult_result=a*b
floor_div_result=a//b
mod_result=a%b

exponent_result=a**b
print("Addition of a and b is:",add_result)
print("Subtraction of a and b is:",sub_result)
print("Multiplication of a and b is:",mult_result)
print("Floor division of a and b is:",floor_div_result) 
print("Modulus of a and b is:",mod_result)
print("Exponent of a and b is:",exponent_result)


```

```python
##comparison operations
a=10    
b=10
print("a==b:",a==b)
print("a!=b:",a!=b)
print("a>b:",a>b)
print("a<b:",a<b)
print("a>=b:",a>=b)
print("a<=b:",a<=b)
```

```python
## logical operations
a=True
b=False
result_and=a and b
result_or=a or b
result_not=not a
print("a and b:",result_and)
print("a or b:",result_or)
print("not a:",result_not)
```
