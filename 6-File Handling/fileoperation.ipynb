```python
## read a whole file
with open("example.txt","r") as file:
    content=file.read()
    print(content)
   
```

```python
## read file line by line
with open("example.txt","r") as file:
    for line in file:
        print(line.strip())## strip() is used to remove the new line character
```

```python
##writing a file
with open("example.txt","w") as file:
    file.write("hello how are you krishna\n")
    file.write("this is new line\n ")
```

```python
## write a file without overwriting
with open ("example.txt","a") as file:
    file.write("hello how are you karan\n")
```

```python
## writing list of lines
lines=["hello how are you krishna\n","this is new line\n","hello how are you karan\n"]
with open("example.txt","a") as file:
    file.writelines(lines)
```

```python
##binary file
##write a binary file
data=b"hello how are you krishna"   
with open("example.bin","wb") as file:
    file.write(data)
```

```python
## reading a binary file
with open("example.bin","rb") as file:
    data=file.read()
    print(data)
```

```python
## read the content from a source file and write to a destination file
with open("example.txt","r") as source_file:
    content=source_file.read()
with open("example_copy.txt","w") as destination_file:
    destination_file.write(content)
```

```python
## read a text file and count the number of lines,words and characters
def count_text_file(file_path):
    with open(file_path,"r") as file:
        lines=file.readlines()
        line_count=len(lines)
        words=[line.split() for line in lines]
        word_count=sum([len(line) for line in words])
        characters=sum([len(line) for line in lines])
    return line_count,word_count,characters
filepath="example.txt"
line_count,word_count,characters=count_text_file(filepath)
print(f'lines:{line_count},words:{word_count},characters:{characters}')
```

```python
## writing and then reading a file
with open("example.txt","w+") as file:
    file.write("hello how are you krishna\n")
    file.write("this is new line\n")
    ## move the file cursor to the beginning
    file.seek(0)
    content=file.read()
    print(content)
```
