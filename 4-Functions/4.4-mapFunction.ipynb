```python
def square(n):
    return n*n
```

```python
## map function

numbers=[1,2,3,4,5]
squares=list(map(square,numbers))
print(squares)
```

```python
## lambda function with map

numbers=[1,2,3,4,5]
squares=list(map(lambda n:n*n,numbers))
print(squares)
```

```python
##map multiple iterables
numbers1=[1,2,3]
numbers2=[4,5,6]
added_numbers=list(map(lambda n1,n2:n1+n2,numbers1,numbers2))
print(added_numbers)
```

```python
##map() to convert a list of strings to integers

string_list=["1","2","3","4","5"]
integer_list=list(map(int,string_list))
print(integer_list)
```

```python
words=["apple","banana","cherry"]
upper_words=list(map(str.upper,words))
print(upper_words)
```

```python
def get_name(name):
    return name["name"]

people=[
    {"name":"John","age":30},
    {"name":"Alice","age":25},
    {"name":"Bob","age":35}
]
names=list(map(get_name,people))
print(names)
```
