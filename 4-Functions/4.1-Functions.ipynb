```python
## variable length arguments
## positional arguments and keyword arguments
def print_numberr(*args):
    for num in args:
        print(num)
print_numberr(1,2,3,4,5,"ganesh")
```

```python
## keyword arguements
def print_numberr(**kwargs):
    for key,value in kwargs.items():
        print(key,value)
print_numberr(a=1,b=2,c=3,d=4,e=5,f="ganesh")
print_numberr(g=1,h=2,i=3,j=4,k=5,l="ganesh")
```

```python
## combined args and kwargs
def print_numberr(*args,**kwargs):
    for key,value in kwargs.items():
        print(key,value)
    for num in args:
        print(num)
print_numberr(1,2,3,4,5,"ganesh",a=1,b=2,c=3,d=4,e=5,f="ganesh")
print_numberr(1,2,3,4,5,"ganesh",g=1,h=2,i=3,j=4,k=5,l="ganesh")
```
