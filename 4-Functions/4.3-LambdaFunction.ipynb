```python
##lambda function
##lambda arguments: expression
##definition of lambda function 
##syntax of lambda function 
add = lambda a,b:a+b
print(add(10,20))
```

```python
def even(n):
    if n%2==0:
        return True
    else:
        return False
even=lambda n:n%2==0
print(even(10))
```

```python
def addition(a,b,c):
    return a+b+c
addition=lambda a,b,c:a+b+c
print(addition(10,20,30))
```

```python
##map() applies a function to each items of a list
##filter()
##reduce()
numbers=[1,2,3,4,5]
def square(n):
    return n*n
squares=list(map(square,numbers))
print(squares)
```
