```python
from nltk.stem import WordNetLemmatizer

lemmatizer = WordNetLemmatizer()

print(lemmatizer.lemmatize("going"))
print(lemmatizer.lemmatize("going", pos="v"))

'''POS - Noun n
    Verb v
    Adjective a
    Adverb r
    Interjection i
    Determiner d
    Preposition p
    Conjunction c'''
  
```

```python
words=["eating","eaten","eat","eats","writing","written","write","writes","programming","programmed","program"]
```

```python
for word in words:
    print(lemmatizer.lemmatize(word))
    print(lemmatizer.lemmatize(word, pos="v"))
    print()
```
