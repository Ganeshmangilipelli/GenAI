```python
## classification problem 
## comments on products are positive or negative
## reviews -->eating , eat , eaten , 
## stemming is a process of reducing words to their root form
words=["eating","eaten","eat","eats","writing","written","write","writes","programming","programmed","program"]
```

```python
###porter stemmer
from nltk.stem import PorterStemmer
words=["eating","eaten","eat","eats","writing","written","write","writes","programming","programmed","program"]
stemming=PorterStemmer()
for word in words:
    print(word,"-->",stemming.stem(word))
```

```python
##regexStemmer
from nltk.stem import RegexpStemmer
reg_stemmer=RegexpStemmer("ing$|s$|es$|ed$|ly$")
reg_stemmer.stem("ingsitting")
```

```python
##snowballStemmer
from nltk.stem import SnowballStemmer
snowballsstemmer=SnowballStemmer("english")
for word in words:
    print(word,"-->",snowballsstemmer.stem(word))
```
