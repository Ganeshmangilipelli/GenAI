```python
## Basic magic methods
class Person:   
    def __init__(self, name, age):
        self.name = name
        self.age = age
    def __str__(self):
        return f"Name: {self.name}, Age: {self.age}"
    def __repr__(self):
        return f"Person('{self.name}', {self.age})"
person=Person("John", 25)
print(person)
print(repr(person))
```
