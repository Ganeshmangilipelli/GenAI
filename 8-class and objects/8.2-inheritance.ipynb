```python
'''inheritance in python is a way to create a new class that is a combination of an existing class and another class.
it allows you to reuse the code and functionality of an existing class in a new class.
it is a way to create a new class that is a specialization of an existing class.'''
class Car:
    def __init__(self, make, model, year):
        self.make = make
        self.model = model
        self.year = year

    def drive(self):
        print(f"The {self.make} {self.model} is driving.")

class ElectricCar(Car):
    def __init__(self, make, model, year, battery_size=75):
        super().__init__(make, model, year)
        self.battery_size = battery_size

    def charge(self):
        print(f"The {self.make} {self.model} is charging .")
        

my_car = ElectricCar("Tesla", "Model S", 2022)
my_car.drive()  # Output: The Tesla Model S is driving.
my_car.charge()  # Output: The Tesla Model S is charging.
```

```python
##multiple inheritance
##when a class inherits from more than one class, it is called multiple inheritance
class Animal:
    def __init__(self, name):
        self.name = name

    def speak(self):
        print(f"{self.name} is speaking.")
class Pet:
    def __init__(self,owner):
        self.owner = owner

class Dog(Animal, Pet):
    def __init__(self, name, owner):
        Animal.__init__(self, name)
        Pet.__init__(self, owner)

    def bark(self):
        print(f"{self.name} is barking.")

my_dog = Dog("Fido", "John")
my_dog.bark()  # Output: Fido is barking.
my_dog.speak()  # Output: Fido is speaking.
my_dog.owner  # Output: John
```
