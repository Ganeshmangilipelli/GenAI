```python
from abc import ABC, abstractmethod
##abstarct class
class Vehicle(ABC):
  
    def drive(self):
        print("The vehicle is driving")
    
    @abstractmethod
    def start_engine(self):
        pass
class Car(Vehicle):
    def start_engine(self):
        print("The car engine is starting")
        
def operate_vehicle(vehicle):
    vehicle.start_engine()
    vehicle.drive()
car = Car()
operate_vehicle(car)
```
