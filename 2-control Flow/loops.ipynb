```python
## strings 
str = "Hello, World!"
for i in str:
    print(i)
```

```python
## while loop
i = 0
while i < 5:
    print(i)
    i += 1
```

```python
count =0
while count % 2 == 0:
    print(count)
    count += 1  
```

```python
## loop control statements
## break statement exits the loop when a condition is met
for i in range(10):
    if i == 3:
        break
    print(i)
```

```python
## continue statement skips the current iteration and continues with the next iteration of the loop
for i in range(10):
    if i % 2== 0:
        continue
    print(i)
```

```python
## pass statement is a null operation; it does nothing when executed. It can be used as a placeholder in loops, functions, or classes where code is syntactically required but you don't want to execute any code.
for i in range(10):
    if i == 3:
        pass
    else:
        print(i)    
```

```python
## nested loops
for i in range(3):
    for j in range(3):
        print(f"i: {i}, j: {j}")
    print(" ")
```

```python
##calculate the sum of first natural numbers using a while loop
n = 10
sum = 0
i = 1
while i <= n:
    sum = sum + i
    i = i + 1
print("The sum is", sum)
```

```python
## using for loop to calculate the sum of first natural numbers
n = 10
sum = 0 
for i in range(1, n + 1):
    sum = sum + i
print("The sum is", sum)
```

```python
## example of prime number using for loop
for num in range(1,101):
    if num > 1:
        for i in range(2, num):
            if (num % i) == 0:
                break
        else:
            print(num)  
```
