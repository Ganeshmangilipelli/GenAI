```python
## if statement
age = 20
if age >= 18:
    print("You are eligible to vote.")
```

```python
## else statement
age = 15
if age >= 18:
    print("You are eligible to vote.")
else:
    print("You are not eligible to vote.")
```

```python
## elif statement
age = 15    
if age >= 18:
    print("You are eligible to vote.")
elif age >= 16:
    print("You are eligible for a driving license.")
else:
    print("You are not eligible to vote or drive.") 
    
```

```python
## Nested condtional statements
age = 20
if age >= 18:
    print("You are eligible to vote.")
    if age >= 21:
        print("You are also eligible to drink alcohol.")
    else:
        print("You are not eligible to drink alcohol.")
else:
    print("You are not eligible to vote.")  
```

```python
## practical example
## determine if a year is a leap year or not
year = 2000
if year % 4 == 0:
    if year % 100 == 0:
        if year % 400 == 0:
            print(year, "is a leap year.")
        else:
            print(year, "is not a leap year.")
    else:
        print(year, "is a leap year.")
else:
    print(year, "is not a leap year.")
```

```python
## simple calculator

num1 = float(input("Enter the first number: "))
num2 = float(input("Enter the second number: "))
operator = input("Enter the operator (+, -, *, /): ")

if operator == "+": 
    result = num1 + num2
    print("Result:", result)
elif operator == "-":
    result = num1 - num2
    print("Result:", result)
elif operator == "*":
    result = num1 * num2
    print("Result:", result)
elif operator == "/":
    if num2 != 0:
        result = num1 / num2
        print("Result:", result)
    else:
        print("Error: Division by zero is not allowed.")
else:
    print("Error: Invalid operator. Please use +, -, *, or /.") 
```

```python
### determine ticket price based on age
age = int(input("Enter your age: "))
if age < 3:
    print("Ticket price: Free")
elif age >= 3 and age <= 12:
    print("Ticket price: $10")
else:
    print("Ticket price: $15")  
```
